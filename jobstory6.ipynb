# Business Understanding

We willen kijken of er jaren zijn waar beter games of juist minder goede games released worden. Om zo verder te kunnen kijken wat er anders was in die jaren om zo een beter inzicht te krijgen in wat een goede game is.

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('appstore_games.csv')
```

# data Understanding

```python
df.info()
```

# data preperation

```python
df1 = df[['Name', 'Average User Rating', 'Original Release Date' ]]

#hier worden de collumns geselecteerd die gebruikt gaan worden
```

```python
df1
```

```python
df1.info()
#kijken of er  nulls aanwezig zijn
```

```python
df1 = df1.dropna()
#alle null values verwijderen
```

```python
df1
```

```python
df1['Original Release Date'] = df1['Original Release Date'].str[-4:]
df1 = df1.sort_values(by='Original Release Date' , ascending = True)
df1
#de original release date wordt omgezet naar alleen het jaar en gesort op jaar.
```

```python
ax = sns.boxplot(x="Original Release Date", y="Average User Rating", data=df1)
# een boxplot van de average user rating per jaar.
```

je kan duidelijk zien dat nieuwe games beter scoren. Ook is te zien dat in de periodo 2015 - 2018 games hoog scoreden met alleen uitschieters naar beneden.

```python
df2 = df1.groupby('Original Release Date', as_index=False)['Average User Rating'].mean()
#het gemiddelde average user rating wordt berekend
```

```python
df2
```

```python
df3 = df1.groupby('Original Release Date').count()
#een count wordt  uitgevoerd om te kijken hoeveel games er gereleased waren in dat jaar die een avg user rating hadden
```

```python
df3.reset_index()
```

```python
df3 = df3.drop(columns='Name')
#drop the column Name omdat deze geen waardevolle info heeft
```

```python
df3 = df3.rename(columns={"Average User Rating": "Games gereleased"})
#rename de column Average user rating naar Games released
```

```python
df3
```

```python
df4 = df2.merge(df3, on='Original Release Date')
```

```python
df4
```

# conclusie

De nieuwere games hebben over het algemeen een hogere average user rating wel is te zien dat 2019 wat meer lagere scores had maar
